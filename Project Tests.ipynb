```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as intg
```

```python
def omega(r,b):
    return 1/(1 + np.exp(b*(1- r)))

rshow = np.linspace(0,2,100)
b = 10
#for b in range(10,11):
plt.plot(rshow,omega(rshow,b),label="b="+str(b))

#plt.legend()
plt.show()
```

```python
b = 10
r = np.linspace(0,2,1000)
integrand = r/omega(r,b)
dr = r[1]
f = np.exp(-np.cumsum(integrand*dr))

plt.plot(r,f)
plt.show()
```

```python
def derr(r,t,b):
    return -omega(r,b)/r
    #if r <= 1:
    #    return -0.05/r
    #else:
    #    return -1/r

b = 10
t0 = 1
ts = np.linspace(t0,10,100000)

tp = np.linspace(0,t0,1000)
rp = np.sqrt(tp)
r0 = rp[-1]

rsol = intg.odeint(derr,r0,ts,args=(b,))

rp = np.matrix(rp)
rtot = np.concatenate((rp.T,rsol),axis=0)
ttot = np.concatenate((tp,ts))


plt.plot(ttot,rtot)
plt.xlabel("t")
plt.ylabel("r")
plt.show()

```
